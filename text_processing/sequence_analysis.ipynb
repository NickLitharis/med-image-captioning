```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'D:\med-image-captioning\data\csv\image_uri_and_captions_100_cleaned.csv')
```

```python
df['caption_length']=df['s2_caption'].apply(lambda x: len(x.split()))
# print(size(df['tokenized_caption'][0]))
```

```python
#make a bar diagram of caption lengths
plt.figure(figsize=(10,5))
plt.hist(df['caption_length'],bins=50)
plt.xlabel('Caption Length')
plt.ylabel('Number of Captions')
plt.title('Caption Length Distribution')
plt.show()
```

```python
#calculate the maximum length of a caption
max_length=max(df['caption_length'])
print(max_length)
```

```python
#calculate the minimum length of a caption
min_length=min(df['caption_length'])
print(min_length)
```

```python
#calculate the average length of a caption
avg_length=sum(df['caption_length'])/len(df['caption_length'])
print(avg_length)
```

```python
#caculate the median length of a caption
median_length=df['caption_length'].median() 
print(median_length)
```

```python
#make a digram of the caption lengths by 10s with a line for the average length and one for the median length and add a line for the average of the average plus the median
plt.figure(figsize=(10,5))
plt.hist(df['caption_length'],bins=50)
plt.axvline(avg_length,color='red',label='Average Length')
plt.axvline(median_length,color='green',label='Median Length')
plt.axvline((avg_length+median_length)/2,color='orange',label='Average of Average and Median')
plt.xlabel('Caption Length')
plt.ylabel('Number of Captions')
plt.title('Caption Length Distribution')
plt.legend()
plt.show()
```

```python
#print the frequency of captions grouping their lengths by tens
print(df['caption_length'].value_counts(bins=10,sort=False))
```
