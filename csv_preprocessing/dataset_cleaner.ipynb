# Cleaning the dataset

```python
import pandas as pd
import string
import re
```

```python
df=pd.read_csv(r"D:\med-image-captioning\data\csv\dataset.csv",encoding='utf-8')
```

```python
# read second column and convert it to list with lowercase
captions = df['s2_caption'].str.lower().tolist()
```

```python
print(captions[0])
```

```python
print(len(captions))
```

### For every caption remove special characters that are not in the alphabet or digits

```python
for i in range(len(captions)):
    captions[i] = re.sub(r'[^a-z0-9]', ' ', captions[i])
```

```python
print(captions[0])
```

### Removing the words 'figure' or 'fig'

```python
for i in range(len(captions)):
    captions[i] = re.sub(r"^(figure|fig)\s+", ' ', captions[i])
```

```python
print(captions[0])
```

### Removing numbers

```python
for i in range(len(captions)):
    captions[i] = re.sub(r'\d', ' ', captions[i])
```

```python
print(captions[0])
```

### Turning 'x ray' to 'xray'

```python
for i in range(len(captions)):
    captions[i] = re.sub(r'x\s+ray', 'xray', captions[i])
```

### Turning 'ct scan' and 'ct' to 'ctscan'

```python
for i in range(len(captions)):
    captions[i] = re.sub(r'ct\s+scan', 'ctscan', captions[i])
    captions[i] = re.sub(r'\bct\b','ctscan', captions[i])
```

### Turning 'mri scan' and 'mri' to mriscan

```python
for i in range(len(captions)):
    captions[i] = re.sub(r'mri\s+scan', 'mriscan', captions[i])
    captions[i] = re.sub(r'\bmri\b','mriscan', captions[i])
```

### Removing single letters

```python
for i in range(len(captions)):
    captions[i] = re.sub(r'\b[a-z]\b', ' ', captions[i])
```

### Removing metrics like 'mm' or 'cm'

```python
for i in range(len(captions)):
    captions[i] = re.sub(r'\bmm\b', '', captions[i])
    captions[i] = re.sub(r'\bcm\b', '', captions[i])
    captions[i] = re.sub(r'\bdm\b', '', captions[i])
```

### Removing "and" and "the" from the begining anf the end

```python
for i in range(len(captions)):
    captions[i] = re.sub(r'^(?:and|the)\s|\s(?:and|the)$', '', captions[i])
```

### Removing whitespaces

```python
for i in range(len(captions)):
    captions[i] = re.sub(r'^\s+|\s+$', '', captions[i])
    captions[i] = re.sub(r'\s{2,}', ' ', captions[i])
```

```python
for i in range(10):
    print('Caption:',i,captions[i])
```

```python
image_id=df['image_id']
```

## Saving to a new file

```python
# save captions to a new csv file
new_dataset=pd.DataFrame({'image_id': image_id, 'caption': captions})
new_dataset.to_csv(r'D:\med-image-captioning\data\csv\dataset_clean.csv', index=False, encoding='utf-8')
```
